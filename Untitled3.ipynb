```python
import scipy.signal as signal
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 8191, 8192, endpoint=False)
```

```python
square = signal.square(2 * np.pi * 4.00 * t) + np.sin(2 * np.pi * 20.00 * t)
```

```python
plt.plot(t, square)
plt.show()
```

```python
g_square = np.gradient(square)
```

```python
plt.plot(t, g_square)
plt.show()
```

```python
laplacian_square = np.gradient(np.gradient(square))
```

```python
plt.plot(t, laplacian_square)
plt.xlim(2000, 4000)
plt.show()
```
