```python
from cycler import cycler

def plot_settings():
    """
    Set plot feature variables to be used throughout.
    """
    custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
             cycler(linestyle=['-', '--', ':', '-.']))

    plt.rc('lines', linewidth=2)
    plt.rc('axes', prop_cycle=custom_cycler)
    
    return custom_cycler
```

```python
import numpy as np
import matplotlib.pyplot as plt
# from cycler import cycler

def plot_time_series(s, t=None):
    """
    Generic single-series time series plotting recipe. 
    :param s: (array) time-variable dependent parameter series.
    :param t: (array) optional, time axis values. Defaults to integers starting at 0
    :param color: (string) optional, color of the plotted trace
    :param scale: (string) optional, scale of the dependent parameter axis. Default is 'linear'
    """
#     custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
#                  cycler(lw=[1, 2, 3, 4]))
    
#     plt.rc('lines', linewidth=4)
#     plt.rc('axes', prop_cycle=default_cycler)
    cc = plot_settings()
    
    series_length = s.shape[1]
    
    if t is None:
        t = numpy.linspace(1, series_length,1)
    
    
    fig, ax0 = plt.subplots(1, 1)
    ax0.plot(t, s)
    ax0.set_prop_cycle(cc)
    ax0.set_xlim(0, series_length)
    ax0.set_xlabel('time')
    ax0.set_ylabel('signal')
    ax0.grid(True)

    fig.tight_layout()
    plt.show()
```

```python
# create two noisy rectangular waves with different bias voltages represented as 16 bit ADC output values
from scipy import signal
import matplotlib.pyplot as plt

def make_multi_square_signals(number_of_signals, offset_value=(2*np.pi)):
    """
    :param number_of_signals: (int) How many replicas with a set offset 
    will be generated of the same waveform.
    """
    a = 2
    b = 4
    # t length in seconds.
    t_length = 10
    # number of intervals.
    t_intervals = 10000
    t = np.linspace(0, 10, t_intervals)
    offsets = np.linspace(0, offset_value, number_of_signals, endpoint=False)
    s = np.transpose([signal.square(a*np.pi*b*t + phi) for phi in offsets]) 

    return s, t
```

```python
def add_noise_to_signal(s):
    """
    Add random noise to the signals to simulate more realistic conditions.
    :param s: signal to be made noisier.
    :return pure + noise: (array) noise-added array with normally distributied noise. 
    """
#     # Adding noise using target SNR

#     # Set a target SNR
#     target_snr_db = 20
#     # Calculate signal power and convert to dB 
#     sig_avg_watts = np.mean(x_watts)
#     sig_avg_db = 10 * np.log10(sig_avg_watts)
#     # Calculate noise according to [2] then convert to watts
#     noise_avg_db = sig_avg_db - target_snr_db
#     noise_avg_watts = 10 ** (noise_avg_db / 10)
#     # Generate an sample of white noise
#     mean_noise = 0
#     noise_volts = np.random.normal(mean_noise, np.sqrt(noise_avg_watts), len(x_watts))
#     # Noise up the original signal
#     y_volts = x_volts + noise_volts

    pure = s
    noise_scalar = 0.01
    noise = noise_scalar * np.random.normal(0, 1, pure.shape)
    return pure + noise
    
```

```python
# plot the two noisy rectangular waves. 
number_of_signals = 5
offset = np.pi / 2
s, t = make_multi_square_signals(number_of_signals, offset)
noisy_s = add_noise_to_signal(s)
plot_time_series(noisy_s,t)
```
